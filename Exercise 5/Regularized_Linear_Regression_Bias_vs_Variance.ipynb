### 1. Regularized Linear Regression

#### Dataset

#### Cost function and gradient

Regularized linear regression has the following cost function:

$$ J(\theta) = \frac{1}{2m} \left( \sum_{i=1}^m \left( h_\theta\left( x^{(i)} \right) - y^{(i)} \right)^2 \right) + \frac{\lambda}{2m} \left( \sum_{j=1}^n \theta_j^2 \right)$$

where $\lambda$ is the regularization parameter which controls the degree of regularization (used to prevent overfitting). The regularization term puts a penalty on the overall cost $J$. As the magnitudes of the model parameters $\theta_j$ increase, the penalty increases as well.

The partial derivative of the cost function for regularized linear regression (gradient) is defined as:

$$
\begin{align}
& \frac{\partial J(\theta)}{\partial \theta_0} = \frac{1}{m} \sum_{i=1}^m \left( h_\theta \left(x^{(i)} \right) - y^{(i)} \right) x_j^{(i)} & \qquad \text{for } j = 0 \\
& \frac{\partial J(\theta)}{\partial \theta_j} = \left( \frac{1}{m} \sum_{i=1}^m \left( h_\theta \left( x^{(i)} \right) - y^{(i)} \right) x_j^{(i)} \right) + \frac{\lambda}{m} \theta_j & \qquad \text{for } j \ge 1
\end{align}
$$

Note: The $\theta_0$ term should not be regularized.

To compute the cost function and the gradient, I will write the function `linearRegCostFunction`.

```python
# Add column of 1's to variable x
# I create another variable so that when I plot the best fit line, the variable x doesn't have added points
x_intercept = np.concatenate([np.ones((m, 1)), x], axis=1)
```

```python
def linearRegCostFunction(x, y, theta, lambda_ = 0.0):
    m = y.size # number of training examples

    # Initializing variable 'grad' in order to be used below
    grad = np.zeros(theta.shape)
    
    h = x.dot(theta.T) # Hypothesis
    
    # Cost function
    J = (1 / (2 * m)) * np.sum((h - y)**2) + (lambda_ / (2 * m)) * np.sum(theta[1:]**2)
    
    # Gradient
    grad = (1 / m) * (h - y).dot(x)
    grad[1:] =  grad[1:] + (lambda_ / m) * theta[1:]
    
    return J, grad
```

#### Learning parameters

After implementing the cost function and the gradient, the next step is to compute the optimal values of the parameters 𝜃. For this, I will create the function `OptimizeTheta` which will use scipy's `optimization.minimize` method to minimize the cost function.

In this part, I will set the regularization parameter 𝜆 to zero. Because the current implementation of linear regression is trying to fit a 2-dimensional 𝜃, regularization will not be incredibly helpful for a 𝜃 of such low dimension. 

In the later parts of the exercise, I will be using polynomial regression with regularization.

```python
def optimizeTheta(linearRegCostFunction, x, y, lambda_ = 0.0, options = {'maxiter': 200}):
    # Initialize theta as vector of zeros
    initial_theta = np.zeros(x.shape[1])

    # Create "shorthand" for the cost function to be minimized
    costFunction = lambda t: linearRegCostFunction(x, y, t, lambda_)

    result = optimize.minimize(costFunction, initial_theta, jac=True, method='TNC', options=options)
    
    theta = result.x # Parameters computed from advanced optimization algorithm
    
    return theta
```

```python
theta = optimizeTheta(linearRegCostFunction, x_intercept, y)
print(theta)
```

#### Plotting best fit line

After computing the optimal parameters 𝜃, it is time to plot the best fit line on the data and see how well it fits.

```python
plt.figure(figsize=(8,5))
plt.plot(x, y, 'ro', ms=10, mec='k')
plt.plot(x, x_intercept.dot(theta.T), 'b-', label='Best fit line')
plt.title('Training data')
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
plt.legend()
plt.grid(True)
```

### 2. Bias-variance

An important concept in machine learning is the bias-variance tradeoff. Models with high bias are not complex enough for the data and tend to underfit, while models with high variance overfit to the training data.

In this part of the exercise, I will plot training and test errors on a learning curve to diagnose bias-variance problems.

#### Learning Curves

I will write the function `learningCurve` to get the training and cross validation set error for different training set sizes. With these error sets, I am able to plot a learning curve which will be useful in debugging the learning algorithm.

The training error for a dataset is defined as:

$$ J_{\text{train}} = \frac{1}{2m} \left[ \sum_{i=1}^m \left(h_\theta \left( x^{(i)} \right) - y^{(i)} \right)^2 \right] $$

**Important note 1**: The training error does not include the regularization term. One way to compute the training error is to use the existing cost function and set $\lambda$ to 0 only when using it to compute the training error and cross validation error.

**Important note 2**: When I am computing the training set error, I need to make sure to compute it on the training subset (i.e., `x[:n,:]` and `y[:n]`) instead of the entire training set. However, for the cross validation error, I should compute it over the entire cross validation set.

```python
def learningCurve(x, y, x_cv, y_cv, lambda_ = 0):
    m = y.size # Number of training examples

    # Initialize these variables to be used below
    error_train = np.zeros(m)
    error_cv = np.zeros(m)

    for i in range(1, m + 1):
        # Computed over different training subsets
        theta_train = optimizeTheta(linearRegCostFunction, x[:i, :], y[:i], lambda_)
        J_train, _ = linearRegCostFunction(x[:i, :], y[:i], theta_train, lambda_ = 0)
        error_train[i - 1] = J_train
        
        # Computed over entire cross validation set
        J_cv, _ = linearRegCostFunction(x_cv, y_cv, theta_train, lambda_ = 0)
        error_cv[i - 1] = J_cv
    
    return error_train, error_cv
```

```python
# Add column of 1's to variables x_cv
x_cv_intercept = np.concatenate([np.ones((x_cv.shape[0], 1)), x_cv], axis=1)

# Compute training and cross validation error sets
error_train, error_cv = learningCurve(x_intercept, y, x_cv_intercept, y_cv)

#Plot learning curve
# np.arange gives evenly spaced values in a range
plt.plot(np.arange(1, m + 1), error_train, np.arange(1, m + 1), error_cv, lw=2)
plt.title('Learning curve for linear regression')
plt.legend(['Training', 'Cross Validation'])
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.axis([0, 13, 0, 150])

print('# Training Examples\tTrain Error\tCross Validation Error')
for i in range(m):
    print('  \t%d\t\t%f\t%f' % (i+1, error_train[i], error_cv[i]))
```

In the learning curve, we can observe that both the training error and cross validation error are high when the number of training examples is increased. This reflects a high bias problem in the model - the linear regression model is too simple and is underfitting the dataset.

I will address this problem by adding polynomial features to the model which means I will be using polynomial regression.

### 3. Polynomial regression

For polynomial regression, the hypothesis has the following form:

$$
\begin{align}
h_\theta(x)  &= \theta_0 + \theta_1 \times (\text{waterLevel}) + \theta_2 \times (\text{waterLevel})^2 + \cdots + \theta_p \times (\text{waterLevel})^p \\
& = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \cdots + \theta_p x_p
\end{align}
$$

By defining $x_1 = (\text{waterLevel})$, $x_2 = (\text{waterLevel})^2$ , $\cdots$, $x_p =
(\text{waterLevel})^p$, we obtain a linear regression model where the features are the various powers of the original feature (waterLevel).

To add more features using the higher powers of the existing feature $x$, I will create the function `polyFeatures`. When a training set $x$ of size $m \times 1$ is passed into the function, the function will return a $m \times d$ matrix `x_poly` ($d$ = degree of polynomial), where column 1 holds the original values of x, column 2 holds the values of $x^2$, column 3 holds the values of $x^3$, and so on.

```python
def polynomialFeatures(x, d):
    # Initializing the variable to be used below
    x_poly = np.zeros((x.shape[0], d))

    for i in range(1, d + 1):
        x_poly[:, i - 1] = x[:, 0]**i

    return x_poly
```

Since I am adding polynomial features, the features will have different magnitudes and so I need to apply normalization in order to scale the features.

```python
def featureNormalization(x):
    mean = np.mean(x, axis=0)
    x_norm = x - mean

    std = np.std(x_norm, axis=0, ddof=1)
    x_norm /= std
    
    return x_norm, mean, std
```

```python
# Degree of polynomial
d = 8

# Map x onto Polynomial Features and Normalization
x_poly = polynomialFeatures(x, d)
x_poly, mean, std = featureNormalization(x_poly)
x_poly = np.concatenate([np.ones((m, 1)), x_poly], axis=1) # Add column of 1's

# Map x_cv onto Polynomial Features and Normalization (using training set 'mean' and 'std')
x_cv_poly = polynomialFeatures(x_cv, d)
x_cv_poly -= mean
x_cv_poly /= std
x_cv_poly = np.concatenate([np.ones((y_cv.size, 1)), x_cv_poly], axis=1) # Add column of 1's

# Map x_test onto Polynomial Features and Normalization (using training set 'mean' and 'std')
x_test_poly = polynomialFeatures(x_test, d)
x_test_poly -= mean
x_test_poly /= std
x_test_poly = np.concatenate([np.ones((y_test.size, 1)), x_test_poly], axis=1) # Add column of 1's
```

#### Learning parameters

After having added the polynomial features and normalized the features on the training, cross validation and test sets, I will proceed to train polynomial regression using the linear regression cost function. This is because the polynomial terms have been turned into features that we can use for linear regression.


I will use the previously written functions `optimizeTheta` to learn the parameters $\theta$ and `learningCurve` to calculate the training and cross validation error sets and then create two plots for polynomial regression with $\lambda = 0$. The first plot will be the best fit line and the second plot will be the learning curve for the polynomial regression.

To plot the best fit line for polynomial regression over the dataset, I will use a helper function called `plotFit`.

```python
def plotFit(polynomialFeatures, min_x, max_x, mean, std, theta, d):
    # I will plot a range slightly bigger than the min and max values to get an idea of how the best fit line will vary 
    # outside the range of the data points
    x = np.arange(min_x - 15, max_x + 25, 0.05).reshape(-1, 1) # Reshapes the array into 1 column only 
    
    # Map x onto Polynomial Features and Normalization
    x_poly = polynomialFeatures(x, d)
    x_poly -= mean
    x_poly /= std

    # Add column of 1's
    x_poly = np.concatenate([np.ones((x.shape[0], 1)), x_poly], axis=1)

    # Plot
    plt.plot(x, np.dot(x_poly, theta), '--', lw=2)
```

```python
lambda_ = 0

# Training set parameters
theta = optimizeTheta(linearRegCostFunction, x_poly, y,
                      lambda_ = lambda_, options = {'maxiter': 55})

# Training and cross validation error sets
error_train, error_cv = learningCurve(x_poly, y, x_cv_poly, y_cv, lambda_)

# Plot best fit line
plt.figure(figsize=(8,5))
plt.plot(x, y, 'ro', ms=10, mew=1.5, mec='k')
plotFit(polynomialFeatures, np.min(x), np.max(x), mean, std, theta, d)
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
plt.title('Polynomial Regression Fit (lambda = %f)' % lambda_)
plt.ylim([-20, 50])
plt.grid(True)

# Plot learning curve
plt.figure(figsize=(8,5))
plt.plot(np.arange(1, 1 + m), error_train, np.arange(1, 1 + m), error_cv)
plt.title('Polynomial Regression Learning Curve (lambda = %f)' % lambda_)
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.axis([0, 13, 0, 100])
plt.legend(['Training', 'Cross Validation'])
plt.grid(True)

print('Polynomial Regression (lambda = %f)\n' % lambda_)
print('# Training Examples\tTrain Error\tCross Validation Error')
for i in range(m):
    print('  \t%d\t\t%f\t%f' % (i+1, error_train[i], error_cv[i]))
```

On the first plot we can see that the polynomial fit is able to follow the datapoints very well, thus having a low training error. However, it drops off at the extremes which is an indicator that the polynomial regression model is overfitting the training data and will not generalize well to new examples.

On the second plot we can see that the training error is very low, essentially stays zero for all numbers of training samples, but the cross validation error is high. There is a gap between the training and cross validation errors, indicating that there is a high variance problem (overfitting).

One way to combat the overfitting (high variance) problem is to add regularization to the model. I will now select a regularization parameter $\lambda$ using a cross validation set.

#### Selecting regularization parameter $\lambda$

Regularization is important in order to prevent overfitting. A model without regularization ($\lambda = 0$) fits the training set well, but does not generalize well to new examples. Conversely, a model with too much regularization ($\lambda = 100$) does not fit the training set and testing set well and we end up having an underfitting problem. So choosing the regularization parameter $\lambda$ well is important to provide a good fit line to the data.

To select the regularization parameter $\lambda$ I will use a cross validation set to evaluate how good each $\lambda$ value is. After selecting the best $\lambda$ value using the cross validation set, I can then evaluate the model on the test set to estimate how well the model will perform on new data. 

I will write the function `validationCurve` to compute the training error and cross validation error. For this I will use the `optimizeTheta` function to train the model using different values of $\lambda$. I will try $\lambda$ in the following range: {0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10}.

```python
def validationCurve(x, y, x_cv, y_cv):
    # Range of selected values for lambda
    lambda_range = [0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10]

    # Initializing these variables to be used below
    error_train = np.zeros(len(lambda_range))
    error_cv = np.zeros(len(lambda_range))

    # For each of the lambda values in the range
    for i in range(len(lambda_range)):
        lambda_ = lambda_range[i]
        theta_train = optimizeTheta(linearRegCostFunction, x, y, lambda_)
        J_train, _ = linearRegCostFunction(x, y, theta_train, lambda_ = 0)
        error_train[i] = J_train
        
        J_cv, _ = linearRegCostFunction(x_cv, y_cv, theta_train, lambda_ = 0)
        error_cv[i] = J_cv

    return lambda_range, error_train, error_cv
```

```python
lambda_range, error_train, error_cv = validationCurve(x_poly, y, x_cv_poly, y_cv)

# Plot validation curve
plt.figure(figsize=(8,5))
plt.plot(lambda_range, error_train, '-o', lambda_range, error_cv, '-o', lw=2)
plt.legend(['Training', 'Cross Validation'])
plt.xlabel('lambda')
plt.ylabel('Error')
plt.grid(True)

print('lambda\t\tTrain Error\tValidation Error')
for i in range(len(lambda_range)):
    print(' %f\t%f\t%f' % (lambda_range[i], error_train[i], error_cv[i]))
```

In this plot, we can see that the best value of $\lambda$ is around 3.

As such, I will plot the best fit line for polynomial regression using the regularization parameter $\lambda = 3$

```python
lambda_ = 3

# Training set parameters
theta = optimizeTheta(linearRegCostFunction, x_poly, y,
                      lambda_ = lambda_, options = {'maxiter': 55})

# Plot best fit line
plt.figure(figsize=(8,5))
plt.plot(x, y, 'ro', ms=10, mew=1.5, mec='k')
plotFit(polynomialFeatures, np.min(x), np.max(x), mean, std, theta, d)
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
plt.title('Polynomial Regression Fit (lambda = %f)' % lambda_)
plt.ylim([-20, 50])
plt.grid(True)
```
