### Dataset

You are given a data set in `ex3data1.mat` that contains 5000 training examples of handwritten digits (This is a subset of the [MNIST](http://yann.lecun.com/exdb/mnist) handwritten digit dataset). The `.mat` format means that that the data has been saved in a native Octave/MATLAB matrix format, instead of a text (ASCII) format like a csv-file. We use the `.mat` format here because this is the dataset provided in the MATLAB version of this assignment. Fortunately, python provides mechanisms to load MATLAB native format using the `loadmat` function within the `scipy.io` module. This function returns a python dictionary with keys containing the variable names within the `.mat` file. 

There are 5000 training examples in `ex3data1.mat`, where each training example is a 20 pixel by 20 pixel grayscale image of the digit. Each pixel is represented by a floating point number indicating the grayscale intensity at that location. The 20 by 20 grid of pixels is “unrolled” into a 400-dimensional vector. Each of these training examples becomes a single row in our data matrix `x`. This gives us a 5000 by 400 matrix `X` where every row is a training example for a handwritten digit image.

$$ x = \begin{bmatrix} - \: (x^{(1)})^T \: - \\ -\: (x^{(2)})^T \:- \\ \vdots \\ - \: (x^{(m)})^T \:-  \end{bmatrix} $$

The second part of the training set is a 5000-dimensional vector `y` that contains labels for the training set. 

#### Importing libraries and data

```python
# Images of digits 20x20 pixels - Each pixel is a feature so 20 x 20 = 400 features
input_layer_size  = 400

# 10 labels, from 1 to 10 (note that "0" is mapped to label 10 in the dataset)
num_labels = 10

#  Training data stored in arrays x, y
data = loadmat('Data/ex3data1.mat')
x, y = data['X'], data['y'].ravel() # flattens the array, putting each value of every row in a single row 
                                    # (this way we get 5000-dimensional vector) 

# Set the zero digit to 0, rather than its mapped 10 in this dataset
# This is due to the fact that this dataset was used in MATLAB where there is no index 0
y[y == 10] = 0

m = y.size # Number of training examples

n = x.shape[1] # Number of features
```

#### Visualizing the data

I will start by visualizing a subset of the training set. I will use the function `displayData` that picks 100 randomly selected rows from `x` and maps each row to a 20 pixel by 20 pixel grayscale image and displays the images together.

```python
indices_to_display = np.random.choice(m, 100, replace=False)
selection = x[indices_to_display, :] # contains 100 random rows from x

def displayData(x, figsize=(10, 10)):
    # Compute rows, cols
    if x.ndim == 2:
        m, n = x.shape
    elif x.ndim == 1: #This allows vectors to be displayed as well
        n = x.size
        m = 1
        x = x[None]  # Promote to a 2 dimensional array
    else:
        raise IndexError('Input X should be 1 or 2 dimensional.')
        
    width, height = 20, 20 # width and height of each digit image (20x20 pixel images)
    display_rows = int(np.floor(np.sqrt(m))) # Plot height
    display_cols = int(np.ceil(m / display_rows)) # Plot width
    
    # ax_array will store all axes objects (each row will have display_cols objects with a total of display_rows)
    # axes objects are where we will display each image
    fig, ax_array = plt.subplots(display_rows, display_cols, figsize=figsize)
    fig.subplots_adjust(wspace=0.025, hspace=0.025) # Creates gaps between the images
    
    #If x is a vector then create list of the vector otherwise flatten the array
    ax_array = [ax_array] if m == 1 else ax_array.ravel()
    
    for i, ax in enumerate(ax_array):
        #In each axis object we will display the image of a digit
        ax.imshow(x[i].reshape(width, height, order='F'), cmap='Greys', extent=[0, 1, 0, 1])
        # order='F' turns the images 90 degrees counter-clockwise, cmap is the color palette for the img and
        # Extent defines the left, right, bottom and top limits
        ax.axis('off') # Turns off the axes for the images
```

Before doing anything else, I will add a column of 1's as the first column of the variable x that correspond to $x_0$.

```python
x = np.concatenate([np.ones((m, 1)), x], axis=1)
```

#### Sigmoid function

I will first start by creating the sigmoid function so that I can reuse it in other parts of my code, making my life easier.

```python
def sigmoid(z):
    g = 1 / (1 + np.exp(-z)) # np.exp calculates the exponential of all values in array

    return g
```

#### Regularized cost function and gradient

Now I will write the function to compute the regularized cost function and gradient using a vectorized approach.

```python
def regCostFunction(theta, x, y, lambda_):
    # Hypothesis
    h = sigmoid(x.dot(theta.T))
    
    # Cost function value
    J = (1 / m) * np.sum(-y.dot(np.log(h)) - (1 - y).dot(np.log(1 - h))) + (lambda_ / (2 * m)) * np.sum(theta[1:]**2)
    
    temp = theta.copy() # Create copy of theta so as to not change anything in theta
    temp[0] = 0 # Since for j=0 we don't regularize the gradient, we can set the first value to 0 and that way the 
                # regularization term is zero for j=0
    
    # Gradient
    grad = (1 / m) * (h - y).dot(x) + (lambda_ / m) * temp
    
    return J, grad
```

#### One-vs-all Classification

I will implement one-vs-all classification by training multiple regularized logistic regression classifiers, one for each of the $K$ classes in the dataset ($K = 10$).

For this I will write a function called `oneVsAll` which will be using scipy's `optimize.minimize` method to train one classifier for each class. This function will return all the classifier parameters in a matrix $\theta \in \mathbb{R}^{K \times (N +1)}$, where each row of $\theta$ corresponds to the learned logistic regression parameters for one class.

```python
def oneVsAll(x, y, num_labels, lambda_):   
    # Set initial theta
    initial_theta = np.zeros(n + 1)
    
    # Set options for minimize
    options = {'maxiter': 50}
        
    # Initializing variable all_theta to be used below
    all_theta = np.zeros((num_labels, n + 1))
   
    for k in range(num_labels):
        res = optimize.minimize(regCostFunction, 
                                initial_theta, 
                                (x, (y == k), lambda_), 
                                jac=True, 
                                method='CG', # Conjugate gradient algorithm
                                options=options)
        all_theta[k] = res.x

    return all_theta
```

```python
lambda_ = 0.1
all_theta = oneVsAll(x, y, num_labels, lambda_)
```

#### 1.4.1 One-vs-all Prediction

As the final step, I will use the trained logistic regression classifiers to predict the digit contained in a given image. For each training example, I will compute the probability that it belongs to each class and pick the class for which the corresponding logistic regression classifier outputs the highest probability and return the class label (0, 1, ..., K-1) as the prediction.

```python
def predictOneVsAll(all_theta, x):  
    p = np.argmax(sigmoid(x.dot(all_theta.T)), axis=1) # Stores the label prediction for each training example

    return p
```

```python
prediction = predictOneVsAll(all_theta, x)
print('Training Set Accuracy: {:.2f}%'.format(np.mean(prediction == y) * 100))
```

The prediction accuracy for this learning algorithm is 95,1% which is very high and shows that it is quite reliable for future predictions.

#### Model representation 

Our neural network is shown in the following figure:

![Neural network](Figures/neuralnetwork.png)

It has 3 layers: an input layer, a hidden layer and an output layer. Our inputs are pixel values of digit images. Since the images are of size 20×20, this gives us 400 input layer units (excluding the extra bias unit which always outputs +1).

I have been provided with a set of network parameters ($\Theta^{(1)}$, $\Theta^{(2)}$) already trained. These are stored in `ex3weights.mat`. The following cell loads those parameters into  `Theta1` and `Theta2`. The parameters have dimensions that are sized for a neural network with 25 units in the second layer and 10 output units (corresponding to the 10 digit classes).

```python
# Setting up the parameters I will use for this exercise
input_layer_size  = 400  # 20x20 Input Images of Digits
hidden_layer_size = 25   # 25 hidden units
num_labels = 10          # 10 labels, from 0 to 9

# Load the .mat file, which returns a dictionary 
weights = loadmat('Data/ex3weights.mat')

# Get the model weights from the dictionary
# Theta1 has size 25 x 401 - Already includes bias unit
# Theta2 has size 10 x 26  - Already includes bias unit
Theta1, Theta2 = weights['Theta1'], weights['Theta2']

# Move the last column of Theta2 into the first position as the first column, due to legacy from MATLAB indexing,
# since the weight file ex3weights.mat was saved based on MATLAB indexing
Theta2 = np.roll(Theta2, 1, axis=0) # Elements that roll beyond the last position are re-introduced at the first position
```

#### Feedforward Propagation and Prediction

Now I will implement feedforward propagation for the neural network. For this I will create a function called `predict_NN` to return the neural network’s prediction. Similar to the one-vs-all classification, the prediction from the neural network will be the label that has the largest probability output $\left( h_\theta(x) \right)_k$.

```python
def predict_NN(Theta1, Theta2, x):
    # Make sure the input has two dimensions
    if x.ndim == 1:
        x = x[None]  # promote to 2-dimensions
    
    m = x.shape[0] # number of training examples in the given variable 'x'
    
    # Add column of 1's to the input layer matrix
    x = np.concatenate([np.ones((m, 1)), x], axis=1)
    
    z2 = x.dot(Theta1.T)
    a2 = sigmoid(z2)
    
    # Add column of 1's to the hidden layer matrix
    a2 = np.concatenate([np.ones((a2.shape[0], 1)), a2], axis=1)
    
    z3 = a2.dot(Theta2.T)
    p = np.argmax(sigmoid(z3), axis=1) # Stores the label prediction for each training example

    # =============================================================
    return p
```

```python
prediction_NN = predict_NN(Theta1, Theta2, x)
print('Training Set Accuracy: {:.1f}%'.format(np.mean(prediction_NN == y) * 100))
```

The accuracy of the neural network is a little higher compared to the accuracy of the One-vs-all classification, being even closer to 100% accuracy.
