## Exercise

You will implement linear regression with one variable to predict profits for a food truck. Suppose you are the CEO of a restaurant franchise and are considering different cities for opening a new outlet. The chain already has trucks in various cities and you have data for profits and populations from the cities. You would like to use this data to help you select which city to expand to next. 

The file `Data/ex1data1.txt` contains the dataset for our linear regression problem. The first column is the population of a city (in 10,000s) and the second column is the profit of a food truck in that city (in $10,000s). A negative value for profit indicates a loss. 

#### Importing libraries and data

```python
# Scientific and vector computation for python
import numpy as np

# Plotting library
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # needed to plot 3-D surfaces

# tells matplotlib to embed plots within the notebook
%matplotlib inline
```

```python
# Read comma separated data
data = np.loadtxt('Data/ex1data1.txt', delimiter=',')
x, y = data[:, 0], data[:, 1]

m = y.size  # number of training examples
```

#### Plotting data

First I am going to take a look at the data to see what I am working with:

```python
plt.figure(figsize=(10,6))
plt.plot(x, y, 'ro', ms=10, mec='k')
plt.grid(True)
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
```

The objective is to minimize the cost function:</center>

$$ J(\theta) = \frac{1}{2m} \sum_{i=1}^m \left( h_{\theta}(x^{(i)}) - y^{(i)}\right)^2$$

where the hypothesis $h_\theta(x)$ is given by the linear model:
$$ h_\theta(x) = \theta^Tx = \theta_0 + \theta_1 x_1$$

To minimize the cost function $J(\theta)$ we have to adjust the parameters $\theta_j$. We do this by using the "batch" gradient descent algorithm. In this algorithm, each iteration performs the update:

$$ \theta_j = \theta_j - \alpha \frac{1}{m} \sum_{i=1}^m \left( h_\theta(x^{(i)}) - y^{(i)}\right)x_j^{(i)} \qquad \text{simultaneously update } \theta_j \text{ for all } j$$

With each step of gradient descent, the parameters $\theta_j$ come closer to the optimal values that will achieve the lowest cost J($\theta$).

### Implementation

First I insert a column of 1's to the variable $x$ as the first column. This column represents the $x_0$ variable.

```python
x = np.stack([np.ones(m), x], axis=1) # Only run this cell once
```

Now I will create a function to compute the cost function:

```python
def computeCostFunction(x, y, theta):
    h = x.dot(theta.T)
    J = (1/(2 * m)) * np.sum((h - y)**2)
    
    return J
```

To make sure it is working properly:

```python
starting_theta = np.zeros(2) #Initializing parameters with value equal to zero

J = computeCostFunction(x, y, starting_theta)
print('%.2f' % J)
```

##### Gradient descent:

```python
def gradientDescent(x, y, theta, alpha, num_iterations):
    # make a copy of theta to avoid changing the original array
    theta = theta.copy()
    
    J_history = [] # Use a list to save cost in every iteration
    theta_history = [] # Use a list to save thetas in every iteartion for plotting later
    for i in range(num_iterations):
        temp = np.zeros(theta.shape)
        theta_history.append(theta)
        for j in range(len(theta)):
            h = x.dot(theta.T)
            temp[j] = theta[j] - alpha * (1 / m) * np.sum((h - y) * x[:, j])
            
        theta = temp
        J_history.append(computeCostFunction(x, y, theta))
    
    return theta, J_history, theta_history

alpha = 0.01 # Learning rate
num_iterations = 1500

theta, J_history, theta_history = gradientDescent(x, y, starting_theta, alpha, num_iterations)

print('Theta found by gradient descent: {:.4f}, {:.4f}'.format(*theta))

# Plot the convergence of the cost function
plt.figure(figsize=(10,6))
plt.plot(list(range(num_iterations)), J_history, '-r')
plt.grid(True)
plt.title("Convergence of Cost Function")
plt.xlabel("Number of iterations")
plt.ylabel("Cost function")
```

With this graph we can check if the alpha we chose is good and that the cost function converges to a mininum value.

To verify if the cost function fits well with the data I am going to plot it on top of the first graph I used to visualize the data:

```python
# plot the linear fit
plt.figure(figsize=(10,6))
plt.plot(x[:, 1], y, 'ro', ms=10, mec='k', label='Training Data')
plt.plot(x[:, 1], np.dot(x, theta), '-', label='Hypothesis: h(x) = %0.2f + %0.2fx'%(theta[0],theta[1]))
plt.grid(True)
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.legend()
```

From what I can see it seems that the implementation of the gradient descent algorithm is correct. Now all that is left to do is to use our hypothesis in order to make some predictions.

```python
# Predict values for population sizes of 35,000 and 70,000 inhabitants
prediction1 = np.dot([1, 3.5], theta)
print('For a population of 35,000 inhabitants, I predict a profit of {:.2f}$\n'.format(prediction1*10000))

prediction2 = np.dot([1, 7], theta)
print('For a population of 70,000 inhabitants, I predict a profit of {:.2f}$\n'.format(prediction2*10000))
```

### Bonus

As an extra I am going to plot the surface and contour plots of the cost function just to visualize it better.

```python
# Grid over which we will calculate J
theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-1, 4, 100)

# Initialize J_vals to a matrix of 0's
J_vals = np.zeros((theta0_vals.shape[0], theta1_vals.shape[0]))

# Fill out J_vals
for i, theta0 in enumerate(theta0_vals):
    for j, theta1 in enumerate(theta1_vals):
        J_vals[i, j] = computeCostFunction(x, y, np.array([theta0, theta1]))
        
# Because of the way meshgrids work in the surf command, we need to
# transpose J_vals before calling surf, or else the axes will be flipped
J_vals = J_vals.T

# Surface plot
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(theta0_vals, theta1_vals, J_vals, cmap='viridis')
plt.xlabel(r'$\theta_0$')
plt.ylabel(r'$\theta_1$')
plt.title('Surface Plot')

# Contour plot
# Plot J_vals as 15 contours spaced logarithmically between 0.01 and 100
ax = plt.subplot(122)
plt.contour(theta0_vals, theta1_vals, J_vals, linewidths=2, cmap='viridis', levels=np.logspace(-2, 3, 20))
plt.plot([x[0] for x in theta_history],[x[1] for x in theta_history],'bo-')
plt.xlabel(r'$\theta_0$')
plt.ylabel(r'$\theta_1$')
plt.plot(theta[0], theta[1], 'ro', ms=10, lw=2)
plt.title('Contour Plot')
```
