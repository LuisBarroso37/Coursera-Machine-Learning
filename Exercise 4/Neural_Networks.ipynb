#### Sigmoid function

I will first start by creating the sigmoid function so that I can reuse it in other parts of my code, making my life easier.

```python
def sigmoid(z):
    g = 1 / (1 + np.exp(-z)) # np.exp calculates the exponential of all values in array

    return g
```

#### Random Initialization

When training neural networks, it is important to randomly initialize the parameters for symmetry breaking. One effective strategy for random initialization is to randomly select values for $\Theta^{(l)}$ in the range $[-\epsilon_{init}, \epsilon_{init}]$. This range of values ensures that the parameters are kept small and makes the learning more efficient.

A good way to choose $\epsilon_{init}$ is to base it on the number of units in the network. A good choice of $\epsilon_{init}$ is $\epsilon_{init} = \frac{\sqrt{6}}{\sqrt{L_{in} + L_{out}}}$, where $L_{in} = s_l$ and $L_{out} = s_{l+1}$ are the number of units in the layers adjacent to $\Theta^{l}$.

To achieve this I will write the function `randInitializeWeights` to initialize the weights for $\Theta$.

```python
import math

def randInitializeWeights(L_in, L_out):
    # Get epsilon_init
    epsilon_init = math.sqrt(6) / math.sqrt(L_in + L_out)

    # Randomly initialize the weights to small values
    W = np.random.rand(L_out, 1 + L_in) * 2 * epsilon_init - epsilon_init

    return W
```

```python
# Initializing neural network parameters
initial_Theta1 = randInitializeWeights(input_layer_size, hidden_layer_size)
initial_Theta2 = randInitializeWeights(hidden_layer_size, num_labels)

# Unroll parameters into vector
initial_nn_params = np.concatenate([initial_Theta1.ravel(), initial_Theta2.ravel()], axis=0)
```

### Cost function and gradient
Now I will implement the cost function and gradient for the neural network. For this purpose I will create the function `nnCostFunction` where I will use feedforward and backpropagation to get the cost function value and the gradients.
#### 1. Feedforward and cost function

I will start by computing the cost function using feedforward to determine the hypothesis $h_\theta \left( x^{(i)} \right)$ as shown in the neural network figure above.

The cost function for the neural network with regularization is:

$$ J(\theta) = \frac{1}{m} \sum_{i=1}^{m}\sum_{k=1}^{K} \left[ - y_k^{(i)} \log \left( \left( h_\theta \left( x^{(i)} \right) \right)_k \right) - \left( 1 - y_k^{(i)} \right) \log \left( 1 - \left( h_\theta \left( x^{(i)} \right) \right)_k \right) \right] + \frac{\lambda}{2 m} \left[ \sum_{j=1}^{25} \sum_{k=1}^{400} \left( \Theta_{j,k}^{(1)} \right)^2 + \sum_{j=1}^{10} \sum_{k=1}^{25} \left( \Theta_{j,k}^{(2)} \right)^2 \right] $$

K = 10 is the total number of possible labels. $h_\theta(x^{(i)})_k = a_k^{(3)}$ is the activation (output value) of the $k^{th}$ output unit. 

For the purpose of training a neural network, the original labels in the variable $y$ (0, 1, ..., 9) need to be encoded as vectors containing only values 0 or 1, so that:

$$ y = 
\begin{bmatrix} 1 \\ 0 \\ 0 \\\vdots \\ 0 \end{bmatrix}, \quad
\begin{bmatrix} 0 \\ 1 \\ 0 \\ \vdots \\ 0 \end{bmatrix}, \quad \cdots  \quad \text{or} \qquad
\begin{bmatrix} 0 \\ 0 \\ 0 \\ \vdots \\ 1 \end{bmatrix}.
$$

For example, if $x^{(i)}$ is an image of the digit 5, then the corresponding $y^{(i)}$ (that I should use with the cost function) should be a 10-dimensional vector with $y_5 = 1$ and the other elements equal to 0.

Note: I should not be regularizing the terms that correspond to the bias units. For the matrices `Theta1` and `Theta2`, this corresponds to the first column of each matrix.

```python
def nnCostFunction(nn_params,
                   input_layer_size,
                   hidden_layer_size,
                   num_labels,
                   x, y, lambda_=0.0):
    
    # Reshape nn_params back into the parameters Theta1 and Theta2, the weight matrices
    # for our 3 layer neural network
    Theta1 = np.reshape(nn_params[:hidden_layer_size * (input_layer_size + 1)],
                        (hidden_layer_size, (input_layer_size + 1)))

    Theta2 = np.reshape(nn_params[(hidden_layer_size * (input_layer_size + 1)):],
                        (num_labels, (hidden_layer_size + 1)))

    # Number of training examples
    m = y.size 
    
    #Initializing these two variables to be used below
    Theta1_grad = np.zeros(Theta1.shape)
    Theta2_grad = np.zeros(Theta2.shape)

    ## Feedforward ##
    #Add column of 1's to variable x
    x = np.concatenate([np.ones((m, 1)), x], axis=1)
     
    z2 = x.dot(Theta1.T)
    a2 = sigmoid(z2)
    
    #Add column of 1's to variable a2
    a2 = np.concatenate([np.ones((a2.shape[0], 1)), a2], axis=1)
    
    z3 = a2.dot(Theta2.T)
    h = sigmoid(z3) #Hypothesis
    ###################
    
    # Turns the values in the variable y (0-9) to the corresponding binary vectors ([1,0,0,0,0,0,0,0,0], ...)
    y_matrix = np.eye(num_labels)[y]
    
    # Regularization term
    reg = (lambda_ / (2 * m)) * (np.sum(Theta1[:,1:]**2) + np.sum(Theta2[:,1:]**2))
    
    # Cost function
    J = (1 / m) * np.sum(-y_matrix * (np.log(h)) - (1 - y_matrix) * (np.log(1 - h))) + reg
    
    ## Back propagation ##
    delta_3 = h - y_matrix
    delta_2 = delta_3.dot(Theta2)[:, 1:] * sigmoidGradient(z2)
    
    Delta_2 = delta_2.T.dot(x)
    Delta_3 = delta_3.T.dot(a2)
    ###################
    
    # Gradients
    Theta1_grad = (1 / m) * Delta_2
    Theta1_grad[:, 1:] = Theta1_grad[:, 1:] + (lambda_ / m) * Theta1[:, 1:]
    Theta2_grad = (1 / m) * Delta_3
    Theta2_grad[:, 1:] = Theta2_grad[:, 1:] + (lambda_ / m) * Theta2[:, 1:]
    
    # Unroll gradients into a vector
    grad = np.concatenate([Theta1_grad.ravel(), Theta2_grad.ravel()])

    return J, grad
```

#### 2. Backpropagation and gradient

After implementing the feedforward and the cost function, it is time to implement the backpropagation algorithm to compute the gradient for the neural network cost function. For this, I will need to update the function `nnCostFunction` so that it returns an appropriate value for `grad`.

I will start by creating a helper function `sigmoidGradient` that calculates the gradient for the sigmoid function.

#### 2.1 Sigmoid Gradient

$$ g'(z) = \frac{d}{dz} g(z) = g(z)\left(1-g(z)\right) $$

where

$$ \text{sigmoid}(z) = g(z) = \frac{1}{1 + e^{-z}} $$

```python
def sigmoidGradient(z):
    g = sigmoid(z) * (1 - sigmoid(z))

    return g
```

#### 2.2 Backpropagation algorithm

![](Figures/ex4-backpropagation.png)

After defining the previous function, now I can implement the backpropagation algorithm.

For backpropagation, first we run feedforward to compute all the activations throughout the network, including the output value of the hypothesis $h_\theta(x)$. Then, for each node $j$ in layer $l$, we compute an “error term” $\delta_j^{(l)}$ that measures how much that node was “responsible” for any errors in our output.

For an output node, we can directly measure the difference between the network’s activation and the true target value, and use that to define $\delta_j^{(3)}$ (since layer 3 is the output layer). For the hidden units, we will compute $\delta_j^{(l)}$ based on a weighted average of the error terms of the nodes in layer $(l+1)$. Below are the steps that we need to do in order to implement backpropagation algorithm (also depicted in the figure above):

1. Set the input layer’s values $(a^{(1)})$ to the $t^{th }$training example $x^{(t)}$. Perform feedforward, computing the activations $(z^{(2)}, a^{(2)}, z^{(3)}, a^{(3)})$ for layers 2 and 3. Note that you need to add a `+1` term to ensure that the vectors of activations for layers $a^{(1)}$ and $a^{(2)}$ also include the bias unit.
<p>&nbsp;</p>

1. For the output units in layer 3 (the output layer), set: 
$$\delta^{(3)} = \left(a^{(3)} - y \right)$$
where $y is the matrix where each row is a vector that represents the label of the training example. 
<p>&nbsp;</p>

1. For the hidden layer $l = 2$, set 
$$ \delta^{(2)} = \left( \Theta^{(2)} \right)^T \delta^{(3)} * g'\left(z^{(2)} \right)$$
Notes: The symbol $*$ performs element wise multiplication in `numpy`. We need to exclude the first column of Theta2 because the hidden layer bias unit has no connection to the input layer so we do not use backpropagation for it.
<p>&nbsp;</p>

1. Accumulate the gradients using the following formula:
$$ \Delta^{(l)} = \delta^{(l+1)} (a^{(l)})^{(T)} $$
<p>&nbsp;</p>

1. Obtain the unregularized gradients for the neural network cost function by dividing the accumulated gradients by $\frac{1}{m}$:
$$ \frac{\partial}{\partial \Theta_{ij}^{(l)}} J(\Theta) = D_{ij}^{(l)} = \frac{1}{m} \Delta_{ij}^{(l)}$$

1. Finally to obtain the regularized gradients, simply add the regularization term to every element of the unregularized gradients except the elements in the first column.
$$ \begin{align} 
&
& \frac{\partial}{\partial \Theta_{ij}^{(l)}} J(\Theta) = D_{ij}^{(l)} = D_{ij}^{(l)} + \frac{\lambda}{m} \Theta_{ij}^{(l)} & \qquad \text{for } j \ge 1
\end{align}
$$

#### 2.3  Gradient checking 

To make sure that the backpropagation algorithm was implemented correctly, we perform gradient checking. This method takes a numerical estimate of the gradient and compares it to the gradient computed with backpropagation. If the relative difference between both is very small (e.g. less than 1e-9) then it means that backpropagaton was correctly implemented.

Let's suppose we have a function $f_i(\theta)$ that computes $\frac{\partial}{\partial \theta_i} J(\theta)$; we’d like to check if $f_i$ is outputting correct derivative values.

$$
\text{Let } \theta^{(i+)} = \theta + \begin{bmatrix} 0 \\ 0 \\ \vdots \\ \epsilon \\ \vdots \\ 0 \end{bmatrix}
\quad \text{and} \quad \theta^{(i-)} = \theta - \begin{bmatrix} 0 \\ 0 \\ \vdots \\ \epsilon \\ \vdots \\ 0 \end{bmatrix}
$$

So, $\theta^{(i+)}$ is the same as $\theta$, except its $i^{th}$ element has been incremented by $\epsilon$. Similarly, $\theta^{(i−)}$ is the corresponding vector with the $i^{th}$ element decreased by $\epsilon$. We can now numerically verify $f_i(\theta)$’s correctness by checking, for each $i$, that:

$$ f_i\left( \theta \right) \approx \frac{J\left( \theta^{(i+)}\right) - J\left( \theta^{(i-)} \right)}{2\epsilon} $$

The degree to which these two values should approximate each other will depend on the details of $J$. But assuming $\epsilon = 10^{-4}$, we’ll usually find that the left- and right-hand sides of the above will agree to at least 4 significant digits (and often many more).

To execute gradient checking I will use 3 functions:

The function `debugInitializeWeights` will initialize the weights of a layer with fan_in incoming connections and fan_out outgoings connections using a fixed strategy. This will help us later in debugging.

The function `computeNumericalGradient` will be used to compute the numerical gradient.

The function `checkNNGradients` will create a small neural network and dataset that will be used for checking the gradients. If the backpropagation implementation is correct, I should see a relative difference that is less than 1e-9.

**Practical Tip:** When performing gradient checking, it is much more efficient to use a small neural network with a relatively small number of input units and hidden units, thus having a relatively small number of parameters. Each dimension of $\theta$ requires two evaluations of the cost function and this can be expensive. In the function `checkNNGradients`, the code creates a small random model and dataset which is used with `computeNumericalGradient` for gradient checking. Furthermore, after I am confident that the gradient computations are correct, I should disable gradient checking before running the learning algorithm.

**Practical Tip:** Gradient checking works for any function where we are computing the cost and the gradient. Concretely, we can use the same `computeNumericalGradient` function to check if our gradient implementations for the other exercises are correct too (e.g. logistic regression’s cost function).

```python
def debugInitializeWeights(fan_out, fan_in):
    #fan_out are the number of outgoing connections and fan_in are the number of incoming connections
  
    # Initialize W using "sin". This ensures that W is always of the same values and will be useful for debugging
    W = np.sin(np.arange(1, 1 + (1 + fan_in) * fan_out)) / 10.0
    W = W.reshape(fan_out, 1 + fan_in, order='F')
    return W
```

```python
def computeNumericalGradient(J, theta, e=1e-4):
    # Initialize variable in order to be used below
    num_grad = np.zeros(theta.shape)
    
    # Creates identity matrix in which every value in the diagonal is epsilon - this ensures that in each iteration 
    # epsilon will be added to the correct index
    epsilon = np.diag(e * np.ones(theta.shape))
    
    for i in range(theta.size):
        loss1, _ = J(theta - epsilon[:, i])
        loss2, _ = J(theta + epsilon[:, i])
        num_grad[i] = (loss2 - loss1) / (2 * e)
    return num_grad
```

```python
def checkNNGradients(nnCostFunction, lambda_=0):
    # Neural network layers and training set
    input_layer_size = 3
    hidden_layer_size = 5
    num_labels = 3
    m = 5

    # Generate some 'random' test data
    Theta1 = debugInitializeWeights(hidden_layer_size, input_layer_size)
    Theta2 = debugInitializeWeights(num_labels, hidden_layer_size)

    # Reusing debugInitializeWeights to generate x
    x = debugInitializeWeights(m, input_layer_size - 1)
    y = np.arange(1, 1 + m) % num_labels
    
    # Unroll parameters into a vector
    nn_params = np.concatenate([Theta1.ravel(), Theta2.ravel()])

    # Shorthand for cost function
    costFunc = lambda p: nnCostFunction(p, input_layer_size, hidden_layer_size,
                                        num_labels, x, y, lambda_)
    
    # Get gradient from backpropagation
    cost, grad = costFunc(nn_params)
    
    # Get numerical gradient
    num_grad = computeNumericalGradient(costFunc, nn_params)

    # Display the two gradient computations
    print(np.stack([num_grad, grad], axis=1))
    print('The above two columns should be very similar.')
    print('(Left - Numerical Gradient, Right - Analytical Gradient)\n')

    # Evaluate the norm of the difference between the two gradient computations. If the implementation is correct,
    # and assuming e = 0.0001 in computeNumericalGradient, then diff should be less than 1e-9.
    diff = np.linalg.norm(num_grad - grad) / np.linalg.norm(num_grad + grad)

    print('If the backpropagation implementation is correct, then \n'
          'the relative difference will be small (less than 1e-9). \n'
          'Relative Difference: %g' % diff)
```

```python
# Check gradients by running checkNNGradients
lambda_ = 3
checkNNGradients(nnCostFunction, lambda_)
```

After applying gradient checking, I see that my implementation of backpropagation is correct and that the function `nnCostFunction` is returning the correct values for the gradient.

Now, after having successfully implemented the neural network cost function and gradient computation, the next step is to use an advanced optimization algorithm to learn a good set of parameters. For this I will use the method `optimize.minimize` from the `scipy` library.

#### Learning parameters

```python
# Setting maximum number of iterations
options = {'maxiter': 400}

# Regularization parameter
lambda_ = 0.03

# Create "short hand" for the cost function to be minimized
costFunction = lambda p: nnCostFunction(p, input_layer_size,
                                        hidden_layer_size,
                                        num_labels, x, y, lambda_)

# Now costFunction is a function that takes in only one argument (the neural network parameters)
result = optimize.minimize(costFunction,
                        initial_nn_params,
                        jac=True,
                        method='TNC',
                        options=options)

# Get the parameters of the optimization
nn_params = result.x
        
# Obtain Theta1 and Theta2 back from nn_params
Theta1 = np.reshape(nn_params[:hidden_layer_size * (input_layer_size + 1)],
                    (hidden_layer_size, (input_layer_size + 1)))

Theta2 = np.reshape(nn_params[(hidden_layer_size * (input_layer_size + 1)):],
                    (num_labels, (hidden_layer_size + 1)))
```

Finally I will write the function `predict` in order to make a prediction of the training set using the neural network and see its accuracy when compared to the values in the variable y. For the prediction I will use feedforward again.

#### Prediction

```python
def predict(Theta1, Theta2, x):
    # Make sure the input has two dimensions
    if x.ndim == 1:
        x = x[None]  # promote to 2-dimensions
    
    m = x.shape[0] # number of training examples in the given variable 'x'
    
    # Add column of 1's to the input layer matrix
    x = np.concatenate([np.ones((m, 1)), x], axis=1)
    
    z2 = x.dot(Theta1.T)
    a2 = sigmoid(z2)
    
    # Add column of 1's to the hidden layer matrix
    a2 = np.concatenate([np.ones((a2.shape[0], 1)), a2], axis=1)
    
    z3 = a2.dot(Theta2.T)
    p = np.argmax(sigmoid(z3), axis=1) # Stores the label prediction for each training example
    
    return p
```

```python
prediction = predict(Theta1, Theta2, x)
print('Training Set Accuracy: %f' % (np.mean(prediction == y) * 100))
```
